## Gates

Here are some _gates_ and how we apply them in PennyLane. Gates are applied inside a circuit function by specifying the qubits (wires) they act on.

```python
# Create a device with 4 qubits
dev_demo = qml.device("default.qubit", wires=4)

@qml.qnode(dev_demo)
def circuit():
    #Apply NOT gate to qubit 2
    qml.PauliX(wires=2)

    #Apply Z gate to qubit 2
    qml.PauliZ(wires=2)

    #Apply H gate to qubit 3
    qml.Hadamard(wires=3)

    #Apply CNOT gate where qubit 2 is control and qubit 0 is target
    qml.CNOT(wires=[2,0])

    #Apply CZ gate where qubit 0 is control and qubit 1 is target
    qml.CZ(wires=[0,1])

    #Apply SWAP gates to qubits 1 and 3
    qml.SWAP(wires=[1,3])

    #Apply Toffoli (CCNOT) gate where qubit 0 and 1 are the control and qubit 2 is the target
    qml.Toffoli(wires=[0,1,2])

    return qml.state()
```

Let's draw our circuit to visualize the operations.

```python
qml.draw_mpl(circuit)()
```

Here is the same circuit rendered using matplotlib:

```python
qml.draw_mpl(circuit)()
```

## More about gates

It is possible to apply a gate to multiple qubits at once.

```python
# Create a new device
dev_demo2 = qml.device("default.qubit", wires=4)

@qml.qnode(dev_demo2)
def circuit2():
    # Apply H gate to all qubits
    for w in range(4):
        qml.Hadamard(wires=w)
    return qml.state()

qml.draw_mpl(circuit2)()
```

`qml.ctrl` function creates the controlled version of a gate.

```python
# Create a controlled version of the Hadamard gate with 3 control qubits
CCCH = qml.ctrl(qml.Hadamard(wires=2), control=[0,1,3])

@qml.qnode(dev_demo2)
def circuit_ctrl():
    # Apply H to all qubits first
    for w in range(4):
        qml.Hadamard(wires=w)
    # Apply the controlled-H gate
    CCCH
    return qml.state()

qml.draw_mpl(circuit_ctrl)()
```

One can define new gates by arithmetic operations as well

```python
ROOTX = qml.pow(qml.PauliX(wires=1), z=0.5)

@qml.qnode(dev_demo2)
def circuit_pow():
    for w in range(4):
        qml.Hadamard(wires=w)
    qml.ctrl(qml.Hadamard(wires=2), control=[0,1,3])
    ROOTX
    return qml.state()

qml.draw_mpl(circuit_pow)()
```

---## Running Circuits

One way to get results from your circuit is to measure and run it for multiple shots. 

Let's first create a simple circuit and measure it. 

```python
import pennylane as qml

# Create a device with 2 qubits and 1000 shots
dev_measure = qml.device("default.qubit", wires=2)

@qml.set_shots(shots=1000)
@qml.qnode(dev_measure)
def circuit():
    # Apply H gate to qubit 0
    qml.Hadamard(wires=0)
    # Measure both qubits
    return qml.counts(wires=[0,1])
```

Now we will run the circuit and take samples. In PennyLane, calling the circuit function executes the simulation.

```python
# Call the circuit function to execute the simulation
result = circuit()

# Get the results as a dictionary
print(result)
```

The output of the circuit are the states $\ket{00}$ and $\ket{01}$ and they are measured almost equal number of times. At this point, we need to make a note about the order of the qubits in PennyLane.

PennyLane uses the following order for qubits:

$$ wires[0], wires[1]$$

For example, when $X$ gate is applied to wire 0, the state $\ket{00}$ becomes $\ket{01}$. If you want to obtain state $\ket{10}$, then you should apply $X$ gate to wire 1.


It is also possible to measure only some of the qubits. Let's only measure qreg[0] this time.

```python
import pennylane as qml

# Create a device with 2 qubits and 1000 shots
dev_measure2 = qml.device("default.qubit", wires=2)

@qml.set_shots(shots=1000)
@qml.qnode(dev_measure2)
def circuit_partial():
    # Apply H gate to qubit 0
    qml.Hadamard(wires=0)
    # Measure only qubit 0
    return qml.counts(wires=[0])

result = circuit_partial()
print(result)
```

## Debugging the circuits - State Representation

It is possible to get the exact quantum state from the simulator. You shouldn't measure your circuit before getting the state. 

We will use qml.state() to obtain the vector representing the quantum state.

```python
import pennylane as qml

# Create a device with 2 qubits (no shots for exact state)
dev_state = qml.device("default.qubit", wires=2)

@qml.qnode(dev_state)
def circuit():
    # Apply H gate to qubit 0
    qml.Hadamard(wires=0)
    return qml.state()

state = circuit()

print(np.round(state, 3))
```

If you use simulator after the measurement, you will observe that the quantum state has collapsed to one of the states.

```python
# Note: In PennyLane, if you add measurements to a state-returning circuit,
# it will raise an error. To demonstrate state collapse, we can use qml.density_matrix()
# or simply note that measurement collapses the state classically.
# Here we show the state without measurement for comparison:
print('State before measurement:', np.round(state, 3))
print('After measurement, the state collapses to either |00> or |01>')
```

---## Unitary Matrix Representation

It is possible to obtain *unitary* matrix representation of gates and circuits.

```python
ROOTX = qml.pow(qml.PauliX(wires=0), z=0.5)

print('Unitary matrix representation of the CNOT gate')
print(np.round(qml.matrix(qml.CNOT(wires=[0,1])), 3))
print('Unitary matrix representation of the ROOTX gate we have created')
print(np.round(qml.matrix(ROOTX), 3))
```

```python
import pennylane as qml

# Create a device with 2 qubits
dev_unitary = qml.device("default.qubit", wires=2)

@qml.qnode(dev_unitary)
def circuit():
    # Apply H gate to all qubits
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)
    return qml.state()

# Get the unitary matrix of the circuit
matrix = qml.matrix(circuit)()

print('Unitary matrix representation of H operator on 2 qubits.')
print(np.round(matrix, 3))
```
